```python
import os
import requests
import random
import math
from getpass import getpass

# Character sets
sp = [chr(i) for i in range(33, 48)] + [chr(i) for i in range(58, 65)] + \
     [chr(i) for i in range(91, 97)] + [chr(i) for i in range(123, 127)]
no = [chr(i) for i in range(48, 58)]
caplet = [chr(i) for i in range(65, 91)]
lowlet = [chr(i) for i in range(97, 123)]

# ------------------------------------------
# Password Quality Checks
# ------------------------------------------

def length(password):
    if len(password) < 8:
        print("⚠️ Too short. Use at least 8 characters.")
        return True
    print("✅ Length is good.")
    return False

def chara(password):
    upper = any(c.isupper() for c in password)
    lower = any(c.islower() for c in password)
    if not upper and not lower:
        print("⚠️ Missing both uppercase and lowercase letters.")
    elif not upper:
        print("⚠️ Add uppercase letters.")
    elif not lower:
        print("⚠️ Add lowercase letters.")
    else:
        print("✅ Good mix of upper & lower case.")
    return not (upper and lower)

def unq(password):
    count = 1
    for i in range(1, len(password)):
        if password[i] == password[i - 1]:
            count += 1
            if count > 3:
                print("⚠️ Too many repeating characters.")
                return True
        else:
            count = 1
    print("✅ No excessive repetition.")
    return False

def load_common_passwords():
    path = "common_passwords.txt"
    if not os.path.exists(path):
        print("🔁 Downloading common passwords list...")
        url = "https://raw.githubusercontent.com/danielmiessler/SecLists/master/Passwords/Common-Credentials/10k-most-common.txt"
        response = requests.get(url)
        with open(path, 'w') as f:
            f.write(response.text)
    with open(path, 'r') as f:
        return set(line.strip().lower() for line in f)

common_passwords = load_common_passwords()

def is_common_password(password):
    if password.lower() in common_passwords:
        print("⚠️ Common password. Easily guessed.")
        return True
    print("✅ Not a commonly used password.")
    return False

def userr(password, user):
    if sum(1 for i in password if i in user) > 4:
        print("⚠️ Too similar to your username.")
        return True
    print("✅ Password is distinct from your username.")
    return False

def namee(password, name):
    if sum(1 for i in password if i in name) > 4:
        print("⚠️ Password too similar to your name.")
        return True
    print("✅ Name isn’t part of the password.")
    return False

def special(password):
    count = sum(1 for i in password if not i.isalnum())
    if count > 1:
        print("✅ Contains special characters.")
        return True
    print("⚠️ Add symbols like !, @, # for strength.")
    return False

def calculate_entropy(password):
    charset = 0
    if any(c.islower() for c in password): charset += 26
    if any(c.isupper() for c in password): charset += 26
    if any(c.isdigit() for c in password): charset += 10
    if any(not c.isalnum() for c in password): charset += 32
    entropy = len(password) * math.log2(charset) if charset > 0 else 0
    print(f"\n🔐 Entropy: {entropy:.2f} bits")
    if entropy >= 75:
        print("✅ Excellent entropy (very strong).")
    elif entropy >= 50:
        print("⚠️ Moderate entropy.")
    else:
        print("❌ Low entropy.")
    return entropy

# ------------------------------------------
# Strong Password Suggestion
# ------------------------------------------

def suggestion():
    print("\n✨ Let's generate a strong password for you!")
    
    while True:
        try:
            length = int(input("🔢 Enter desired password length (min 8): "))
            if length < 8:
                print("❌ Too short! Please enter at least 8.")
                continue
            break
        except ValueError:
            print("❌ Please enter a valid number.")

    characters = sp + no + caplet + lowlet

    # Ensure a mix of character types
    password = [
        random.choice(sp),
        random.choice(no),
        random.choice(caplet),
        random.choice(lowlet)
    ]

    # Fill the rest randomly
    password += random.choices(characters, k=length - 4)
    random.shuffle(password)
    final_password = ''.join(password)

    print(f"\n🔐 Your strong {length}-character password is:\n👉 {final_password}")

# ------------------------------------------
# Main Password Strength Checker
# ------------------------------------------

def main():
    name = input("Enter your name: ")
    username = input("Enter your username: ")
    password = getpass(f"Hi {name}, enter your password (input hidden): ")

    print("\n🔍 Checking your password...\n")
    score = 0
    if not length(password): score += 1
    if not chara(password): score += 1
    if not unq(password): score += 2
    if not is_common_password(password): score += 2
    if not userr(password, username): score += 1
    if not namee(password, name): score += 1
    if special(password): score += 2

    print(f"\n🔒 Password score: {score}/10")
    if score == 10:
        print("🎉 Strong password! Great job.")
    elif score >= 6:
        print("👍 Decent password, can improve.")
    else:
        print("⚠️ Weak password! Consider changing it.")

    calculate_entropy(password)

# ------------------------------------------
# Main Menu Loop
# ------------------------------------------

def menu():
    while True:
        print("\n" + "="*50)
        print("🔐 Welcome to the Password Strength Checker")
        print("1️⃣  Check your password strength")
        print("2️⃣  Suggest a strong password")
        print("3️⃣  Exit")
        print("="*50)
        try:
            ch = int(input("Choose an option: "))
            if ch == 1:
                main()
            elif ch == 2:
                suggestion()
            elif ch == 3:
                print("👋 Goodbye! Stay secure.")
                break
            else:
                print("❌ Invalid choice. Try 1, 2, or 3.")
        except ValueError:
            print("❌ Please enter a valid number.")

# Run it
if __name__ == "__main__":
    menu()
```
